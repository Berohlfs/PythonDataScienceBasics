```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Generate 100 random numbers between 1 and 10
data = np.random.randint(1, 11, 100)
# data = np.random.normal(1, 11, 100) // use this if you want normal dist
df = pd.DataFrame({'Value': data})
```

```python
df['Value'].plot(kind='kde', title='Data with Mean, Median, and Mode', legend=True, color='blue', label='Value')
plt.show()
```

### Measures of Central Tendency
Central tendency refers to the idea of identifying the central or typical value for a probability distribution.

**Mean**: The mean is the average of the numbers. It is calculated by summing all the numbers and then dividing by the count of those numbers.

**Median**: The median is the middle number in a sorted list of numbers. If there's an even number of observations, it's the average of the two middle numbers.

**Mode**: The mode is the number that appears most frequently in a set of numbers.

```python
mean = df['Value'].mean()
median = df['Value'].median()
mode = df['Value'].mode().iloc[0]
print(f"Mean: {mean}, Median: {median}, Mode: {mode}")
df.head()
```

```python
df['Value'].plot(kind='kde', title='Data with Mean, Median, and Mode', legend=True, color='blue', label='Value')
plt.axvline(x=mean, color='red', linestyle='--', label='Mean')
plt.axvline(x=median, color='green', linestyle='--', label='Median')
plt.axvline(x=mode, color='orange', linestyle='--', label='Mode')

plt.legend()
plt.show()
```

**The following code demonstrates a positively skewed data**

```python
# Generate random numbers between 0 and 1
data = np.random.rand(100)

# Apply a square root transformation to introduce positive skew
skewed_data = np.sqrt(data)

# Scale the numbers to a desired range, e.g., between 1 and 10
scaled_data = skewed_data * 100 // 1

df = pd.DataFrame({'Value': scaled_data})


mean = df['Value'].mean()
median = df['Value'].median()
mode = df['Value'].mode().iloc[0]
print(f"Mean: {mean}, Median: {median}, Mode: {mode}")

df['Value'].plot(kind='kde', title='Data with Mean, Median, and Mode', legend=True, color='blue', label='Value')
plt.axvline(x=mean, color='red', linestyle='--', label='Mean')
plt.axvline(x=median, color='green', linestyle='--', label='Median')
plt.axvline(x=mode, color='orange', linestyle='--', label='Mode')

plt.legend()
plt.show()
```

Variance and Standard Deviation are fundamental concepts in statistics, often used to measure the spread or dispersion of a dataset.

#### Variance
Variance measures how far each number in the set is from the mean and thus from every other number in the set.
#### Standard Deviation
Standard Deviation is the square root of the variance. It is a measure of the amount of variation or dispersion in a set of values.

```python
import seaborn as sns
import pandas as pd

# Load an example dataset from seaborn
df = sns.load_dataset('tips')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Calculate variance and standard deviation for 'total_bill'
variance = df['total_bill'].var()
std_dev = df['total_bill'].std()

print(f"Variance of Total Bill: {variance}")
print(f"Standard Deviation of Total Bill: {std_dev}")
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(df['total_bill'], kde=True)
plt.axvline(df['total_bill'].mean(), color='r', linestyle='--', label='Mean')
plt.axvline(df['total_bill'].mean() + std_dev, color='g', linestyle='-', label='Mean + 1 Std Dev')
plt.axvline(df['total_bill'].mean() - std_dev, color='g', linestyle='-', label='Mean - 1 Std Dev')
plt.legend()
plt.title('Distribution of Total Bills with Mean and Standard Deviation')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```
