```python
import pandas as pd

# Sample data
data = {
    'Height (m)': [1.65, 1.70, 1.75, 1.80, 1.68, 1.72, 1.78, 1.85, 1.60, 1.74, 1.82, 1.88],
    'Weight (kg)': [55, 62, 80, 75, 60, 65, 70, 60, 50, 67, 78, 85]
}

# Create DataFrame
df = pd.DataFrame(data)

df
```

```python
# Calculate variance for each column
variance = df.var()

# Calculate standard deviation for each column
std_deviation = df.std()

# Display the results
print("Variance:\n", variance)
print("\nStandard Deviation:\n", std_deviation)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns


# Create scatter plot with regression line
sns.regplot(x=df['Height (m)'], y=df['Weight (kg)'], scatter=True, color='blue', line_kws={"color": "red"})
plt.title('Scatter plot of Height vs Weight with Regression Line')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.grid(True)
# Annotate variance and standard deviation on the plot
plt.annotate(f"Variance:\nHeight: {variance['Height (m)']:.2f}\nWeight: {variance['Weight (kg)']:.2f}", 
             xy=(0.05, 0.95), xycoords='axes fraction', fontsize=10, verticalalignment='top')
plt.annotate(f"Std Deviation:\nHeight: {std_deviation['Height (m)']:.2f}\nWeight: {std_deviation['Weight (kg)']:.2f}", 
             xy=(0.05, 0.75), xycoords='axes fraction', fontsize=10, verticalalignment='top')

plt.show()
```

```python
# Calculate covariance between the two columns
covariance = df['Height (m)'].cov(df['Weight (kg)'])

print("\nCovariance between 'Height (m)' and 'Weight (kg)':", covariance)
```

```python
# Calculate correlation coefficient between the two columns
correlation = df['Height (m)'].corr(df['Weight (kg)'])

print("\nCorrelation coefficient between 'Height (m)' and 'Weight (kg)':", correlation)

# The value will range between -1 and 1, where 1 indicates a perfect
# positive correlation, -1 indicates a perfect negative correlation,
# and 0 indicates no correlation.
```
